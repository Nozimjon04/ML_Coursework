```python
#Load Raw Dataset
import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw/contract_awards_in_investment_project_financing_since_fy_2020_12-04-2025.csv")

df.head()
```

```python
#Remove Duplicate Rows

df = df.drop_duplicates()
df.shape
```

```python
#Fill missing categorical values with "Unknown"

categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols] = df[categorical_cols].fillna("Unknown")
```

```python
#Fill missing numeric values with median

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
```

```python
#convert Signing Date into year & month

df['Contract Signing Date'] = pd.to_datetime(df['Contract Signing Date'], errors='coerce')

df['Signing Year'] = df['Contract Signing Date'].dt.year
df['Signing Month'] = df['Contract Signing Date'].dt.month
```

```python
#drop the original date column

df = df.drop(columns=['Contract Signing Date'])
```

```python
#Save the Cleaned Dataset

df.to_csv("../data/processed/contracts_clean.csv", index=False)
```
